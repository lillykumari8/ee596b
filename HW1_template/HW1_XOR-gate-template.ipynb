```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

<h2>Define Hyper-parameters</h2>

```python
NUM_ITER = 1000
lr = 0.1
```

<h2>Define your input data and expected labels</h2>

```python
x = np.array([[0, 0], [1, 0], [0, 1], [1, 1]], np.float32)
# one hot encoder labels
y = np.array([[1,0],  [0,1],  [0,1], [1,0]], np.float32)
y_no_ohe = np.array([0,1,1,0])
```

```python
#remember to reset the graph
tf.reset_default_graph()
```

<h2>Define placeholder and Variables</h2>

```python
X = tf.placeholder(tf.float32, shape=[None, 2], name='X')
Y = tf.placeholder(tf.float32, shape=[None, 2], name='Y')


weights = {
    'W1': tf.Variable(tf.random_uniform([2, 2], -1, 1), name='W1'),
    'Wout': tf.Variable(tf.random_uniform([2, 2], -1, 1), name='Wout')
}

biases = {
    'b1': tf.Variable(tf.zeros(shape=[2]), name='b1'),
    'bout': tf.Variable(tf.zeros(shape=[2]), name='bout')
}
```

<h2>Define the neural network</h2>

```python
def neural_net(x):
    layer1_out = tf.nn.sigmoid(tf.add(tf.matmul(x, weights['W1']), biases['b1']))
    out = tf.nn.softmax(tf.add(tf.matmul(layer1_out, weights['Wout']), biases['bout']))
    return out
```

<h2>Define cost function</h2>

```python
#predicted labels
logits = neural_net(X)

#define loss
# loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=logits,labels=Y),name='loss')

loss = -tf.reduce_sum(Y * tf.log(logits), name='loss')
#define optimizer
optimizer = tf.train.GradientDescentOptimizer(learning_rate=lr)
train_op = optimizer.minimize(loss)


#compare the predicted labels with true labels
correct_prediction = tf.equal(tf.argmax(logits,1), tf.argmax(Y,1))

#compute the accuracy by taking average
accuracy = tf.reduce_mean(tf.cast(correct_prediction,tf.float32),name='accuracy')

pred_output = tf.argmax(logits, 1)
#Initialize the variables
init = tf.global_variables_initializer()
```

<h2>Execute training</h2>

```python
with tf.Session() as sess:
    sess.run(init)
    
    for i in range(NUM_ITER):
        #run optimization
        sess.run(train_op, feed_dict={X:x, Y:y})
        
        acc = sess.run(accuracy,feed_dict={X:x, Y:y})
        print("Epoch "+str(i)+", Accuracy= {:.3f}".format(acc))
    
    print("Training finished!")
    print("Testing Accuracy:", sess.run(accuracy, feed_dict={X:x, Y:y}))
    for j in range(x.shape[0]):
        print (x[j], '---->' , pred_output.eval({X: x})[j])
    
    
    xs = np.linspace(-0.2, 1.2)
    ys = np.linspace(-0.2, 1.2)
    pred = []
    for xi in xs:
        for yi in ys:
            predicted = sess.run(logits, feed_dict={X: [[xi, yi]]})
            pred.append((xi, yi, predicted.argmax()))
    xs_p, ys_p = [], []
    xs_n, ys_n = [], []
    for xi, yi, c in pred:
        if c == 1:
            xs_p.append(xi)
            ys_p.append(yi)
        else:
            xs_n.append(xi)
            ys_n.append(yi)
    plt.figure(figsize=(15, 15), dpi=100)
    plt.plot(xs_p, ys_p, 'yo', xs_n, ys_n, 'ro')
    plt.scatter(x[:, 0], x[:, 1], c=np.squeeze(y_no_ohe), s=200, cmap=matplotlib.colors.ListedColormap(['blue', 'black']))
    plt.show()

    
    W1 = np.squeeze(sess.run(weights['W1']))
    b1 = np.squeeze(sess.run(biases['b1']))
    W2 = np.squeeze(sess.run(weights['Wout']))
    b2 = np.squeeze(sess.run(biases['bout']))
```

<h2>Plot</h2>

```python
#plot out the decision line determined by learned weight and bias
import matplotlib.pyplot as plt
import matplotlib

plot_x = np.array([np.min(x[:, 0] - 0.2), np.max(x[:, 1]+0.2)])
plot_y = 1 / W1[1][0] * (-W1[0][0] * plot_x - b1[0])


plot_x1 = np.array([np.min(x[:, 0] - 0.2), np.max(x[:, 1]+0.2)])
plot_y1 = 1 / W1[1][1] * (-W1[0][1] * plot_x - b1[1])

plt.scatter(x[:, 0], x[:, 1], c=np.squeeze(y_no_ohe), s=100, cmap=matplotlib.colors.ListedColormap(['blue', 'black']))
plt.plot(plot_x, plot_y, color='red', linewidth=2)
plt.plot(plot_x1, plot_y1, color='red', linewidth=2)

plt.xlim([-0.2, 1.2]); plt.ylim([-0.2, 1.2]);
plt.show()
```
